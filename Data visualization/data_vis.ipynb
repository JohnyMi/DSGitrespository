Vytváření grafů pomocí vlastních předdefinovaných funkcí

```python
def plot_timeseries(axis, x, y, color, xlabel, ylabel):
  axis.plot(x, y, color)
  axis.set_xlabel(xlabel)
  axis.set_ylabel(ylabel, color=color)
  axis.tick_params('y', colors=color, labelcolor=color)
```

```python
fig,ax = plt.subplots()

plot_timeseries(
    axis=ax,
    x=merged.index,
    y=merged.Trend,
    color='r',
    xlabel='Time',
    ylabel='CO2'
)

ax2=ax.twinx()
plot_timeseries(
    axis=ax2,
    x=merged.index,
    y=merged.AverageTemperature,
    color='b',
    xlabel='Time',
    ylabel='Temperature'
)
```

Sloupcové grafy
Sloupcové grafy jsou dalším oblíbeným typem vizualizace, který se používá k zobrazení kategoriálních dat pomocí obdélníkových sloupců.

Chcete-li vytvořit sloupcový graf, použijte funkci plt.bar(). Zadejte kategoriální hodnoty jako souřadnice osy X a odpovídající číselné hodnoty jako výšku sloupců.

Sloupcové grafy se běžně používají k zobrazení výsledků průzkumů, údajů o prodeji nebo jakýchkoli údajů s odlišnými kategoriemi. Matplotlib poskytuje flexibilitu při přizpůsobování vzhledu sloupcových grafů, což umožňuje vytvářet vizuálně atraktivní a informativní vizualizace.

```python
url='https://raw.githubusercontent.com/matzim95/ML-datasets/master/population.csv'

population = pd.read_csv(url, index_col='City')
population.head()
```

```python
population_by_sex = population.groupby('Sex')
both, female, male = [population_by_sex.get_group(x) for x in population_by_sex.groups]
```

```python
female = population[population.Sex == 'Female']
male = population[population.Sex == 'Male']
both = population[population.Sex == 'Both Sexes']
```

```python
# all together
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(both.index,both['Value'])

ax.xaxis.set_tick_params(rotation=90)
ax.set_ylabel('Population')
plt.show()
```

Pokud máte více kategorií k porovnání, můžete vykreslit seskupené nebo skládané sloupcové grafy zadáním příslušných hodnot parametrům width a bottom funkce plt.bar().

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(male.index, male['Value'], label='Male')
ax.bar(female.index, female['Value'], label='Female', bottom=male['Value'])

ax.xaxis.set_tick_params(rotation=90)
ax.set_ylabel('Population')

ax.legend()
plt.show()
```

Bodový graf (korelační diagram)
Bodový graf je typ grafu, který zobrazuje vztah mezi dvěma číselnými proměnnými. Jednotlivé datové body znázorňuje jako body na dvourozměrném souřadnicovém systému.

Zde je příklad, jak vytvořit bodový graf:

```python
try:
    # Pokúsi sa zobraziť prvých pár riadkov dataframe 'merged', ak už existuje
    merged.head()
except NameError:
    # Ak dataframe 'merged' neexistuje (je vyvolaná chyba NameError), načítajú sa potrebné súbory

    # Načítanie súboru s údajmi o CO2 z URL, dátum sa nastaví ako index
    url = 'https://raw.githubusercontent.com/matzim95/ML-datasets/master/co2.csv'
    co2 = pd.read_csv(url, parse_dates=True, index_col='Date')

    # Načítanie súboru s údajmi o teplote z URL, dátum sa nastaví ako index (stĺpec 'dt')
    url = 'https://raw.githubusercontent.com/matzim95/ML-datasets/master/temperature.csv'
    temperature = pd.read_csv(url, parse_dates=True, index_col='dt')

    # Premenovanie indexu z 'dt' na 'Date' v dataframe 'temperature', aby sa dal ľahšie spojiť s 'co2'
    temperature.index.names = ['Date']

    # Filtrovanie údajov o teplote pre mesto Paríž
    paris = temperature[temperature.City == 'Paris']

    # Spojenie dátových rámcov 'co2' a filtrovaného dataframe 'paris' na základe dátumu
    # Použije sa vnútorné spojenie ('inner join'), ktoré ponechá len spoločné dátumy
    merged = pd.merge(co2, paris, how='inner', on='Date')
```

```python
# Vytvorenie grafu s jedným podgrafom (figura a os)
fig, ax = plt.subplots()

# Vytvorenie bodového grafu (scatter plot)
# Osa x: údaje o CO2 ('Interpolated'), osa y: priemerná teplota ('AverageTemperature')
ax.scatter(merged['Interpolated'], merged['AverageTemperature'])

# Nastavenie popisu pre os x (CO2 úrovne)
ax.set_xlabel('CO2')

# Nastavenie popisu pre os y (teplota)
ax.set_ylabel('Temperature')

# Zobrazenie grafu
plt.show()
```

```python
def scatter_plt(merged, x_col, y_col, color_col, xlabel, ylabel, cmap='viridis'):
    """
    Vytvorí bodový graf (scatter plot) s farebným rozlíšením podľa zadaného stĺpca.
    
    Parameters:
    - merged: Pandas DataFrame, ktorý obsahuje údaje
    - x_col: Názov stĺpca pre os X (napr. CO2)
    - y_col: Názov stĺpca pre os Y (napr. Priemerná teplota)
    - color_col: Názov stĺpca na farebné rozlíšenie bodov (napr. Roky)
    - xlabel: Popis osi X
    - ylabel: Popis osi Y
    - cmap: Farebná schéma pre body (default: viridis)
    """
    
    # Vytvorenie grafu
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Bodový graf s farebným rozlíšením podľa zadaného stĺpca
    scatter = ax.scatter(
        merged[x_col],  # Hodnoty na osi X
        merged[y_col],  # Hodnoty na osi Y
        c=merged[color_col],  # Farebné rozlíšenie podľa zvoleného stĺpca
        cmap=cmap  # Farebná mapa
    )
    
    # Nastavenie popisov osí
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    # Pridanie farebnej škály
    fig.colorbar(scatter, ax=ax)
    
    # Zobrazenie grafu
    plt.show()

# Zavolanie funkcie pre existujúci dataset
scatter_plt(merged, 'Interpolated', 'AverageTemperature', 'Year', 'CO2', 'Temperature')
```

Navíc rozšiřuje možnosti knihovny Matplotlib o další statistické funkce. Obsahuje funkce, jako jsou regresní graf, korelační heatmapa, kategoriální grafy, vizualizace časových řad a podporu statistických anotací.
Integrace s Pandas: bezproblémově se integruje s populární knihovnou Pandas pro manipulaci s daty. Jako vstup může přímo přijímat Pandas DataFrame, což je vhodné pro práci s tabulkovými daty.
Atraktivní výchozí styly: obsahuje estetické výchozí styly a palety barev.
Seaborn je celkově výkonná knihovna pro vizualizaci dat, zejména v oblasti statistické grafiky. Doplňuje Matplotlib a Pandas a usnadňuje vytváření krásných a informativních vizualizací s minimálním úsilím.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from IPython.display import display
pd.plotting.register_matplotlib_converters()
```

```python
url = 'https://raw.githubusercontent.com/matzim95/ML-datasets/master/wine.csv'
wine = pd.read_csv(url)
wine.head()
```

Bodový graf s regresním modelem
Největší výhodou používání aplikace Seaborn je množství funkcí pro tvorbu grafů. Například vytvoření bodového grafu v kombinaci s regresním modelem je pouze jeden řádek kódu pomocí funkce sns.lmplot().

To lze provést tímto způsobem:

Předáním objektu DataFrame do argumentu data a zároveň předáním názvů sloupců do argumentů x a y.

```python
sns.lmplot(x='Alcohol', y='Magnesium', data=wine)
plt.show()
```

Seaborn ve skutečnosti nemá speciální funkci pro vykreslení bodového grafu. K vytvoření tohoto grafu použijeme funkci Seaborn sns.lmplot(), která nám umožňuje sestrojit a vykreslit regresní přímku.

Každá funkce vykreslování má naštěstí několik užitečných možností, které lze nastavit. Zde se dozvíte, jak můžeme funkci lmplot() přizpůsobit vašim potřebám:

nastavte fit_reg = False, abyste odstranili regresní přímku, protože chceme bodový graf.
nastavte hue = 'Class', aby se naše body zbarvily podle třídy příkladu.

```python
sns.lmplot(
    x='Alcohol',
    y='Magnesium',
    data=wine,
    fit_reg=False,
    hue='Class'
)

plt.show()
```

ÚKOL PRO VÁS
Pomocí knihovny seaborn vytvořte graf rozptylu znázorňující vztah mezi koncentrací alkoholu a intenzitou barvy:

jako kategorii zadejte třídu vína
zahrňte lineární regresní přímku pro každou třídu vína zvlášť
naformátujte graf
přidejte popisy os, název,
změňte natočení hodnot na ose X,
přidejte do grafu mřížku.

```python
sns.lmplot(
    x='Alcohol',
    y='Color intensity',
    data=wine,
    fit_reg=True,
    hue='Class'
)
plt.xlabel('Alcohol')
plt.ylabel('Color intensity')
plt.title('Alcohol vs color dependence')
plt.xticks(rotation=40)


plt.show()
```
