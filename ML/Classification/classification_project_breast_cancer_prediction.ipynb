```python
import pandas as pd
```

```python
#Pozor na index_col - pri zobrazení matice loc[0] aby nezobrazovalo vzťahy voči indexu
data= pd.read_csv('breast-cancer.csv', index_col="id")
```

```python
# premena objekt dtype na 0 a 1 -> pokial diagnosis stlpec obsahuje B - Benign -> 0, M - Malignant -> 1
data["diagnosis"] = data["diagnosis"].apply(lambda x: 0 if x == 'B' else 1)
```

```python
corr = data.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# treshold, od kdy beru jako relevantní prediktor, pozor !!! míra lineární závislosti (může nemusí)
corr_treshold = 0.4
# vyfiltruj sloupce, jejichž korelace na cílovou proměnou je větší než treshold
interesting_cols = abs(corr.iloc[[0]]) > corr_treshold
filter_cols = [data.columns[i] for i in range(len(data.columns)) if interesting_cols.iloc[0].values[i] == True]
```

```python
# data pouze se sloupci nad tresholdem pro korelaci
data_corr = data[filter_cols]
```

```python
data_corr
```

```python
from sklearn.metrics import roc_curve, auc  # Import správnych funkcií

# Výpočet ROC křivky a AUC pro první model
fpr1, tpr1, _ = roc_curve(y_test, y_pred)
roc_auc_logreg = auc(fpr1, tpr1)

# Výpočet ROC křivky a AUC pro druhý model
fpr2, tpr2, _ = roc_curve(y_test, y_pred_knn)
roc_auc_knn = auc(fpr2, tpr2)

# Výpočet ROC křivky a AUC pro třetí model
fpr3, tpr3, _ = roc_curve(y_test, y_gs_knn)
roc_auc_gs_knn = auc(fpr3, tpr3)

# Výpočet ROC křivky a AUC pro čtvrtý model
fpr4, tpr4, _ = roc_curve(y_test, y_pred_svc)
roc_auc_svc = auc(fpr4, tpr4)

# Vykreslení ROC křivek
plt.figure()
plt.plot(fpr1, tpr1, color='blue', lw=2, label=f'LogReg (AUC = {roc_auc_logreg:.2f})')
plt.plot(fpr2, tpr2, color='red', lw=2, label=f'KNN (AUC = {roc_auc_knn:.2f})')
plt.plot(fpr3, tpr3, color='green', lw=2, label=f'GS KNN (AUC = {roc_auc_gs_knn:.2f})')
plt.plot(fpr4, tpr4, color='yellow', lw=2, label=f'SVC (AUC = {roc_auc_svc:.2f})')

# Přidání 45° čáry (náhodné hádání)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2)

# Nastavení os a legendy
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Multiple Models')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc  # Import správnych funkcií

# Výpočet ROC křivky a AUC pro první model
fpr1, tpr1, _ = roc_curve(y_test, y_pred)
roc_auc_logreg = auc(fpr1, tpr1)

# Výpočet ROC křivky a AUC pro druhý model
fpr2, tpr2, _ = roc_curve(y_test, y_pred_knn)
roc_auc_knn = auc(fpr2, tpr2)

# Výpočet ROC křivky a AUC pro tretí model
fpr3, tpr3, _ = roc_curve(y_test, y_gs_knn)
roc_auc_gs_knn = auc(fpr3, tpr3)

# Výpočet ROC křivky a AUC pro štvrtý model
fpr4, tpr4, _ = roc_curve(y_test, y_pred_svc)
roc_auc_svc = auc(fpr4, tpr4)

# Vykreslení ROC křivek
plt.figure(figsize=(8, 6))  # Nastavenie veľkosti grafu
plt.plot(fpr1, tpr1, color='blue', lw=2, label=f'LogReg (AUC = {roc_auc_logreg:.2f})')
plt.plot(fpr2, tpr2, color='red', lw=2, label=f'KNN (AUC = {roc_auc_knn:.2f})')
plt.plot(fpr3, tpr3, color='green', lw=2, label=f'GS KNN (AUC = {roc_auc_gs_knn:.2f})')
plt.plot(fpr4, tpr4, color='yellow', lw=2, label=f'SVC (AUC = {roc_auc_svc:.2f})')

# Přidání 45° čáry (náhodné hádání)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2)

# Nastavení os a legendy
plt.xlim([0.0, 1.0])  # Osa x od 0.0 do 1.0
plt.ylim([0.9, 1.0])  # Osa y od 0.8 do 1.0
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Multiple Models (Zoomed In)')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
