```python
# Step 2 : import data
import pandas as pd

salary = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
salary.columns
```

```python
y = salary['Salary']
X = salary[['Experience Years']]
```

```python
sns.scatterplot(x='Experience Years', y='Salary', data=salary)
```

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=2529)
```

# check shape of train and test sample

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# Step 6 : train or fit model
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

# Step 8 : model accuracy

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, root_mean_squared_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
root_mean_squared_error(y_test, y_pred)
```

```python
# Step 1: Scatterplot of the actual data points
import matplotlib.pyplot as plt

sns.scatterplot(x=X_test['Experience Years'], y=y_test, color='blue', label='Actual')

# Step 2: Plot the predicted regression line
sns.lineplot(x=X_test['Experience Years'], y=y_pred, color='red', label='Predicted')

# Step 3: Adding labels and title to the plot
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Scatterplot with Regression Line')

# Step 4: Display the plot
plt.legend()
plt.show()
```

```python
# Step 1: Calculate residuals (errors)
import matplotlib.pyplot as plt
import numpy as np

errors = y_test - y_pred

# Step 2: Scatterplot of the actual data points
sns.scatterplot(x=X_test['Experience Years'], y=y_test, color='blue', label='Actual')

# Step 3: Plot the predicted regression line
sns.lineplot(x=X_test['Experience Years'], y=y_pred, color='red', label='Predicted')

# Step 4: Add error bars to the plot (Residuals as error bars)
plt.errorbar(
    X_test['Experience Years'],
    y_test,
    yerr=np.abs(errors),
    fmt='o',
    color='blue',
    ecolor='gray',
    alpha=0.5,
    capsize=5,
    label='Error bars',
)


# Step 5: Adding labels and title to the plot
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Scatterplot with Regression Line and Error Bars')

# Step 6: Display the plot
plt.legend()
plt.show()
```

4. Náměty na vylepšení
Feature Engineering: Můžeš zkusit přidat nové prvky, pokud je máš k dispozici (např. počet místností, lokalitu) nebo například vytvářet polynomiální rozšíření proměnné area pro zachycení nelineárních vztahů.

Tvé řešení je tedy velmi dobře zpracované, a pokud máš zájem o další rozšíření nebo pokročilé modely (např. polynomiální nebo regularizované modely), dej vědět!

Polynomiální funkce mohou pomoci, když lineární vztah nestačí k tomu, aby model dobře vystihl data. Použitím PolynomialFeatures ze scikit-learn rozšíříme náš model tak, aby zahrnoval vyšší mocniny proměnné area, což může zlepšit přesnost, pokud je mezi rozlohou a cenou nelineární vztah.

Zde je krok za krokem, jak na to:

1. Importovat a vytvořit polynomiální vlastnosti
Nejprve nastavíme polynomiální vlastnosti a použijeme je na trénovací a testovací data.

```python
from sklearn.preprocessing import PolynomialFeatures

# Nastavíme stupeň polynomu, např. 2 pro kvadratickou funkci
poly = PolynomialFeatures(degree=2)

# Vytvoříme nové vlastnosti pro X_train a X_test
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

2. Trénink modelu s polynomiálními vlastnostmi
Vytvoříme nový regresní model, který bude trénován na polynomiálně rozšířených datech.

```python
# Vytvoříme a natrénujeme model na polynomiálních vlastnostech
model_poly = LinearRegression()
model_poly.fit(X_train_poly, y_train)
```

3. Predikce a vyhodnocení modelu
Po natrénování modelu predikujeme ceny domů v testovací sadě a vyhodnotíme jeho přesnost.

```python
# Predikce na testovacích datech
y_pred_poly = model_poly.predict(X_test_poly)

# Výpočet metrik pro vyhodnocení
mae_poly = mean_absolute_error(y_test, y_pred_poly)
mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = model_poly.score(X_test_poly, y_test)

print(f"MAE: {mae_poly}")
print(f"MSE: {mse_poly}")
print(f"R^2: {r2_poly}")
```

4. Vizualizace výsledků
Nakonec můžeme vizualizovat polynomiální regresní křivku oproti skutečným hodnotám.

```python
import matplotlib.pyplot as plt


import numpy as np


import seaborn as sns


# Rozsah hodnot `area` pro hladkou křivku

area_range = np.linspace(X_test['area'].min(), X_test['area'].max(), 100).reshape(-1, 1)
area_range_poly = poly.transform(area_range)

price_pred_range = model_poly.predict(area_range_poly)


# Scatter plot skutečných dat a polynomiální křivky

plt.figure(figsize=(10, 6))

sns.scatterplot(x=X_test['area'], y=y_test, color='blue', label="Skutečné hodnoty")
sns.lineplot(x=area_range.flatten(), y=price_pred_range, color='red', label="Polynomiální křivka")

plt.xlabel('Rozloha (area)')
plt.ylabel('Cena (price)')
plt.title("Polynomiální regrese")
plt.legend()
plt.show()
```
